```python
import numpy as np
from scipy import linalg
from numpy.random import default_rng

def singular_approximation(A, rank):
  m, n = A.shape
  V, s, Uh = linalg.svd(A)
  V.shape,  s.shape, Uh.shape

  sigma = np.zeros((m, n))
  for i in range(min(m, n, rank)):
      sigma[i, i] = s[i]
  B = V @ sigma @ Uh
  return B
```
