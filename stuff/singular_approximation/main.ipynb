```python
def print_to_rank(image, r):
  n = r + 1
  height = 2
  width = (n + height - 1) // height

  fig = plt.figure(figsize=(25, 20))
  fig.add_subplot(height, width, 1, xticks=[], yticks=[])
  plt.imshow(image)
  plt.title(f"original\nrank = {rank(image)}");
  for i in range(n):
    index = i + 1
    fig.add_subplot(height, width, width * (index % height) + (index // height + 1), xticks=[], yticks=[])
    plt.imshow(singular_approximation(image, i))
    plt.title(f"rank = {i}");

  plt.show()
```

```python
images = get_images()
for image in images:
  print_to_rank(image, 20)
  print()
  print()
```
